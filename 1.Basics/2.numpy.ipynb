### How to access elements from a array 

```python
a = np.array([100,120,130,190,200,14,18,26])
```

```python
#positive indexing
a[1]
```

```python
a[-1] #negative indexing
```

```python
#BASED ON SLICING
a[0:5]
```

```python
a[0:5:2]
```

```python
#ACCESSING ELEMENT USING INDEX NUMBERS
a[[0,1,5,6,4,0,0,0,0]]
```

```python
#ACCESSING ELEMENTS BASED ON CONDITION
a = np.array([45,46,47])

a[a>45]
```
