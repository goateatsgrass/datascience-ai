### Array
- it is  datatype
- it is a homogeneous elements can only be stored
- it is more strict , same datatype only

it is a part of numpy , array will be n-dimensional

```python
import numpy as np # importing the library
array_n1  =  np.array([1,2,3,4])
```

```python
print(array_n1)
print(type(array_n1))
print(type(array_n1[0]))
```

```python
array_n1  =  np.array([1,2,3,4],dtype= "int16") #change the datatype
array_n1  =  np.array([1,2,3,4],dtype= np.int16)
```

```python
print(array_n1)
print(type(array_n1))
print(type(array_n1[0]))
```

```python
 #change the datatype
array_n2  =  np.array([1,2,3,4],dtype= np.float16)
print(type(array_n2[0]))
```
